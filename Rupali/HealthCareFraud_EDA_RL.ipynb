```python
# changing gender to 0 and 1
train_beneficiary['Gender']=train_beneficiary['Gender'].map({1:1,2:0})
```

```python
# changing RenalDiseaseIndicator to 0 and 1
train_beneficiary['RenalDiseaseIndicator']=train_beneficiary['RenalDiseaseIndicator'].map({"0":0,"Y":1})
```

```python
# changing all chronic diseases to 0 and 1
train_beneficiary['ChronicCond_Alzheimer']=train_beneficiary['ChronicCond_Alzheimer'].map({2:0,1:1})
train_beneficiary['ChronicCond_Heartfailure']=train_beneficiary['ChronicCond_Heartfailure'].map({2:0,1:1})
train_beneficiary['ChronicCond_KidneyDisease']=train_beneficiary['ChronicCond_KidneyDisease'].map({2:0,1:1})
train_beneficiary['ChronicCond_Cancer']=train_beneficiary['ChronicCond_Cancer'].map({2:0,1:1})
train_beneficiary['ChronicCond_ObstrPulmonary']=train_beneficiary['ChronicCond_ObstrPulmonary'].map({2:0,1:1})
train_beneficiary['ChronicCond_Depression']=train_beneficiary['ChronicCond_Depression'].map({2:0,1:1})
train_beneficiary['ChronicCond_Diabetes']=train_beneficiary['ChronicCond_Diabetes'].map({2:0,1:1})
train_beneficiary['ChronicCond_IschemicHeart']=train_beneficiary['ChronicCond_IschemicHeart'].map({2:0,1:1})
train_beneficiary['ChronicCond_Osteoporasis']=train_beneficiary['ChronicCond_Osteoporasis'].map({2:0,1:1})
train_beneficiary['ChronicCond_rheumatoidarthritis']=train_beneficiary['ChronicCond_rheumatoidarthritis'].map({2:0,1:1})
train_beneficiary['ChronicCond_stroke']=train_beneficiary['ChronicCond_stroke'].map({2:0,1:1})
```

```python
train_beneficiary['ChronicCond_stroke'].value_counts()
```

```python
# creating new column Age from DOB and DOD
train_beneficiary['DOB']=pd.to_datetime(train_beneficiary['DOB'])
train_beneficiary['DOD']=pd.to_datetime(train_beneficiary['DOD'])
train_beneficiary['Age']=round(((train_beneficiary['DOD'] - train_beneficiary['DOB']).dt.days)/365)
train_beneficiary['Age']=train_beneficiary['Age'].fillna(round(((pd.to_datetime('2009-12-01') - train_beneficiary['DOB']).dt.days)/365))
```

```python
# creating new column whetherDead to flag dead or alive
train_beneficiary.loc[train_beneficiary.DOD.isna(),'WhetherDead']=0
train_beneficiary.loc[train_beneficiary.DOD.notna(),'WhetherDead']=1
```

```python
# dropping DOD and DOB after creating Age
train_beneficiary= train_beneficiary.drop(['DOB','DOD'], axis =1)
```

```python
train_beneficiary['Age']=train_beneficiary['Age'].astype(int)
```

```python
train_beneficiary['WhetherDead']= train_beneficiary['WhetherDead'].astype(int)
```

```python
# creating new column chronic_conditions giving total of all chronic conditions
train_beneficiary['chronic_conditions']=\
train_beneficiary['ChronicCond_Alzheimer']+ train_beneficiary['ChronicCond_Heartfailure']+\
train_beneficiary['ChronicCond_KidneyDisease']+ train_beneficiary['ChronicCond_Cancer']+\
train_beneficiary['ChronicCond_ObstrPulmonary'] + train_beneficiary['ChronicCond_Depression']+\
train_beneficiary['ChronicCond_Diabetes'] + train_beneficiary['ChronicCond_IschemicHeart'] +\
train_beneficiary['ChronicCond_Osteoporasis'] + train_beneficiary['ChronicCond_rheumatoidarthritis']+\
train_beneficiary['ChronicCond_stroke'] 
```

```python
# dropping all the columns used to create chronic_conditions
train_beneficiary = train_beneficiary.drop(['ChronicCond_Alzheimer','ChronicCond_Heartfailure', 'ChronicCond_KidneyDisease',
                   'ChronicCond_Cancer', 'ChronicCond_ObstrPulmonary','ChronicCond_Depression', 'ChronicCond_Diabetes',
                   'ChronicCond_IschemicHeart', 'ChronicCond_Osteoporasis','ChronicCond_rheumatoidarthritis', 'ChronicCond_stroke'],axis = 1)
```
